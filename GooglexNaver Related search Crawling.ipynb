## 구글 연관검색어 크롤링

```python
from bs4 import BeautifulSoup
from selenium import webdriver

def related_words(keyword):
    url = 'https://www.google.com/search?q='+keyword+'&oq='+keyword+'&aqs=chrome.0.69i59j69i60j69i61j69i60.4368j0j7&sourceid=chrome&ie=UTF-8'

    dr = webdriver.Chrome('C:/Users/user/Desktop/chromedriver')
    dr.get(url)

    dr_html = dr.page_source

    soup = BeautifulSoup(dr_html, 'html.parser')

    rel_words = soup.find_all('div', attrs={'class':'s75CSd ZJjDdf LTmh7d'})
    rel_words_text = []
    for word in rel_words:
        rel_words_text.append(word.text)
    
    print(rel_words_text)
    
    dr.close()
    
    return rel_words_text
    
    
```

```python
rel_words_text = related_words('헷갈리는 분리수거')

rel_words_list = []
rel_words_list.append(rel_words_text)

for i in range(len(rel_words_text)):
    rel_words_list.append(related_words(rel_words_text[i]))
```

```python
rel_words_text = related_words('어려운 분리수거')

rel_words_list.append(rel_words_text)

for i in range(len(rel_words_text)):
    rel_words_list.append(related_words(rel_words_text[i]))
    
```

```python
rel_words = ""

for i in range(len(rel_words_list)):
    for j in range(len(rel_words_list[i])):
        rel_words += rel_words_list[i][j] +"\n"
        
google_tokens = [t for t in rel_words.split()]
google_tokens
```

## 네이버 연관검색어 크롤링

```python
from bs4 import BeautifulSoup
from selenium import webdriver

def related_words(keyword):
    url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query='+keyword

    dr = webdriver.Chrome('C:/Users/user/Desktop/chromedriver')
    dr.get(url)

    dr_html = dr.page_source

    soup = BeautifulSoup(dr_html, 'html.parser')

    rel_words = soup.find_all('div', attrs={'class':'tit'})
    rel_words_text = []
    for word in rel_words:
        rel_words_text.append(word.text)
    
    print(rel_words_text)
    
    dr.close()
    
    return rel_words_text
```

```python
rel_words_text = related_words('올바른 분리수거')

rel_words_list = []
rel_words_list.append(rel_words_text)

for i in range(len(rel_words_text)):
    rel_words_list.append(related_words(rel_words_text[i]))
```

```python
rel_words_text = related_words('헷갈리는 분리수거')

rel_words_list.append(rel_words_text)

for i in range(len(rel_words_text)):
    rel_words_list.append(related_words(rel_words_text[i]))
    
```

```python
rel_words = ""

for i in range(len(rel_words_list)):
    for j in range(len(rel_words_list[i])):
        rel_words += rel_words_list[i][j] +"\n"
        
naver_tokens = [t for t in rel_words.split()]
naver_tokens
```

## 키워드 분석

```python
from konlpy.tag import Komoran
from collections import Counter
from wordcloud import WordCloud
from pprint import pprint
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
komoran = Komoran()

tokens = google_tokens + naver_tokens
stop_words = {'분리수거', '분리', '수거', '배출', '방법', '법', '분리배출한다.', '버릴', '때', '버리기', 'OTHER', '분리배출표시' ,'ppt' ,'일러스트', '재활용', 'For', 'you', '함께', '클릭한', '상품', '추천', '쓰레기'}
keywords=[word for word in tokens if not word in stop_words]
keywords
```

```python
poses=[]
for i in range(len(keywords)):
    pos = komoran.pos(keywords[i])
    for n in range(len(pos)):
        poses.append(pos[n])
poses
```

```python
counter = Counter(poses)
counter = {
    word:freq for word, freq in counter.items()
    if word[1][:1] == 'N' or word[1] == "SL"
}

pprint(sorted(counter.items(), key=lambda x:-x[1]))
```

```python
komoran_userdic = Komoran(userdic='./komoran_dict.txt')
keywords = ' '.join(keywords)
user_nouns = komoran_userdic.nouns(keywords)
user_nouns
```

```python
user_count = Counter(user_nouns)
user_count
```

```python
font_path = 'C:\\Windows\\fonts\\경기천년제목_Light.ttf' 

wordcloud = WordCloud(
    background_color='white',
    font_path=font_path,
    width = 800,
    height = 800
)

wordcloud = wordcloud.generate_from_frequencies(user_count)
```

```python
%matplotlib inline

fig = plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
#fig.savefig('wordcloud.png')
```
